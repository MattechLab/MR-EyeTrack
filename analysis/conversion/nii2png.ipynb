```python
import os
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt

# === Path containing the NIfTI files ===
input_dir = "/home/debi/jaime/repos/MR-EyeTrack/results/bids/derivatives/reconstructions/sub-001/anat/"

# Coordinates
x, y, z = [141, 140, 115]

# Output directory (same as input)
output_dir = input_dir

# Loop through matching files
for fname in os.listdir(input_dir):
    if fname.endswith(".nii.gz"):
        fpath = os.path.join(input_dir, fname)
        print(f"Processing {fname}...")

        # Load NIfTI
        img = nib.load(fpath)
        data = img.get_fdata()

        nx, ny, nz = data.shape
        if not (0 <= x < nx and 0 <= y < ny and 0 <= z < nz):
            print(f"âš ï¸ Skipping {fname}: coordinates out of bounds ({data.shape})")
            continue

        # Extract slices
        axial = data[:, :, z]
        coronal = data[:, y, :]
        sagittal = data[x, :, :]

        # Apply transformations
        axial = np.fliplr(np.rot90(axial, 3))
        coronal = np.fliplr(np.rot90(coronal, 3))
        sagittal = np.rot90(sagittal, 3)

        # Save slices
        slices = [(axial, "1"), (coronal, "2"), (sagittal, "3")]
        for slc, suffix in slices:
            plt.figure(figsize=(5, 5))
            plt.imshow(slc, cmap="gray", origin="lower")
            plt.axis("off")
            outname = fname.replace(".nii.gz", f"{suffix}.png")
            outpath = os.path.join(output_dir, outname)
            plt.savefig(outpath, bbox_inches="tight", pad_inches=0)
            plt.close()

        print(f"âœ… Saved axial/coronal/sagittal PNGs for {fname}")
```
